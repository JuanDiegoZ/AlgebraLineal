## Conclucion.
Para que la multiplicación de matrices esté definida el número de columnas de la primera debe coincidir con el número de filas de la segunda.

### Producto interno entre matrices y cuando esta definido.

El producto interno entre 2 matrices esta definido cuando las filas de la primera matriz tiene el mismo tamaño que las columnas de la segunda matriz.

```python
D = np.array([[2,3,3],[4,5,6],[7,8,9],[4,5,7]])
E = np.array([[1,2],[3,4],[5,6]])

print(D.shape)
print(E.shape)

print(D.dot(E))
```

### Propiedades de las matrices

El producto interno de **++matrices ++**es:

Asociativa: Sí Ax(BxC) = (AxB) x C

Distributiva: Sí Ax(B+C) = (AxB) x (AxC)

Conmutativa: NO BxC = CxB

El producto interno de **++vectores ++**es:

Asociativa: Sí Ax(BxC) = (AxB) x C

Distributiva: Sí Ax(B+C) = (AxB) x (AxC)

Conmutativa: Sí BxC = CxB

```python
A = np.array([[2,3,3,6],[4,5,6,6],[7,8,9,7],[4,5,7,8]])
B = np.array([[1,2,3],[3,4,4],[5,6,4],[6,5,7]])
print(A.dot(B))
print(B.dot(A))
```
